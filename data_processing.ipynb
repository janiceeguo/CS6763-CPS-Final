## Selecting from csv

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from collections import Counter
import re
```

```python
df = pd.read_csv('/content/UQ_AAS21_full.csv')
```

```python
def parse_average_rating(rating_str):
    if pd.isna(rating_str) or rating_str == 'no rating':
        return 0.0
    try:
        return float(rating_str.split(' ')[0])
    except ValueError:
        return 0.0

df['Average Rating'] = df['Average Rating'].apply(parse_average_rating)

df['Number of Rating'] = df['Number of Rating'].astype(str).str.replace(',', '', regex=False)
df['Number of Rating'] = pd.to_numeric(df['Number of Rating'], errors='coerce').fillna(0).astype(int)

print("Data types after conversion:")
df.dtypes
```

```python
df["popularity_score"] = df["Number of Rating"] * df["Average Rating"]

def assign_popularity_tier(n):
    if n == 0:
        return "unrated"
    elif n < 10:
        return "very_low"
    elif n < 100:
        return "low"
    elif n < 1000:
        return "medium"
    else:
        return "high"

df["popularity_tier"] = df["Number of Rating"].apply(assign_popularity_tier)
```

```python
def stratified_sample(df, n_per_category=50):
    samples = []
    categories = df["Category"].unique()

    for cat in categories:
        subset = df[df["Category"] == cat]

        if len(subset) < n_per_category:
            samples.append(subset)
        else:
            samples.append(subset.sample(n=n_per_category, random_state=42))

    return pd.concat(samples)
```

```python
def sample_by_popularity(df, n_per_group=50):
    samples = []

    for (cat, tier), group in df.groupby(["Category", "popularity_tier"]):
        if len(group) > n_per_group:
            samples.append(group.sample(n=n_per_group, random_state=42))
        else:
            samples.append(group)

    return pd.concat(samples)
```

```python
# Tokenize invocation names
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df["Invocation Name"])

# Compute token frequency
token_counts = np.array(X.sum(axis=0)).flatten()
tokens = vectorizer.get_feature_names_out()

token_freq = dict(zip(tokens, token_counts))

# Score "genericity" of each invocation name
def genericity_score(text):
    words = text.split()
    if not words:
        return 0
    return np.mean([token_freq.get(w, 0) for w in words])

df["genericity_score"] = df["Invocation Name"].apply(genericity_score)

# Short name indicator
df["length"] = df["Invocation Name"].apply(lambda x: len(x.split()))

# Ambiguity score
df["ambiguity_score"] = (
    df["genericity_score"] * 0.7 +
    (1 / (df["length"] + 1)) * 0.3
)

df = df.reset_index(drop=True)

df["norm_ambiguity_score"] = (
    df["ambiguity_score"] - df["ambiguity_score"].min()
) / (
    df["ambiguity_score"].max() - df["ambiguity_score"].min()
)

def ambiguity_sample(df, n_per_category=50):
    samples = []
    categories = df["Category"].unique()

    for cat in categories:
        subset = df[df["Category"] == cat]

        if len(subset) < n_per_category:
            samples.append(subset)
        else:
            top = subset.sort_values("ambiguity_score", ascending=False)
            samples.append(top.head(n_per_category))

    return pd.concat(samples)
```

```python
strat_sample = stratified_sample(df, n_per_category=40)
pop_sample = sample_by_popularity(df, n_per_group=15)
ambiguous_sample = ambiguity_sample(df, n_per_category=25)
```

```python
final_df = pd.concat([
    strat_sample,
    pop_sample,
    ambiguous_sample
]).drop_duplicates(subset=["Skill Path"]).reset_index(drop=True)

final_df.shape
```
